# Model Training Code

```python
import pandas as pd
```

```python
model_training_dataset = pd.read_csv("../datasets/device_data/final_dataset_for_training_ML_model.csv", index_col=[0])
```

```python
model_training_dataset.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model_training_dataset_copy = model_training_dataset.copy()
```

```python
# Separate features (X) and target variables (y)
X = model_training_dataset.iloc[:, :-3]  # Features (19 columns), 1st dimension is for rows, second one is for columns
y_nitrogen = model_training_dataset["nitrogen_concentration"]
y_phosphorus = model_training_dataset["phosphorus_concentration"]
y_potassium = model_training_dataset["potassium_concentration"]
```

```python
X
```

```python
y_nitrogen
```

```python
y_phosphorus
```

```python
y_potassium
```

```python
# Encode categorical target variables
y_nitrogen.replace({"low": 0, "moderate": 1, "high": 2}, inplace=True)
y_phosphorus.replace({"low": 0, "moderate": 1, "high": 2}, inplace=True)
y_potassium.replace({"low": 0, "moderate": 1, "high": 2}, inplace=True)
```

```python
y_nitrogen
```

```python
y_phosphorus
```

```python
y_potassium
```

```python
# Split data into training and testing sets
X_train, X_test, y_train_nitrogen, y_test_nitrogen, y_train_phosphorus, y_test_phosphorus, y_train_potassium, y_test_potassium = train_test_split(X, y_nitrogen, y_phosphorus, y_potassium, test_size=0.1, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train_nitrogen
```

```python
y_test_nitrogen
```

```python
# Train separate models for each target variable
model_nitrogen = RandomForestClassifier()
model_nitrogen.fit(X_train, y_train_nitrogen)

model_phosphorus = RandomForestClassifier()
model_phosphorus.fit(X_train, y_train_phosphorus)

model_potassium = RandomForestClassifier()
model_potassium.fit(X_train, y_train_potassium)
```

```python
# Evaluate models
y_pred_nitrogen = model_nitrogen.predict(X_test)
accuracy_nitrogen = accuracy_score(y_test_nitrogen, y_pred_nitrogen)

y_pred_phosphorus = model_phosphorus.predict(X_test)
accuracy_phosphorus = accuracy_score(y_test_phosphorus, y_pred_phosphorus)

y_pred_potassium = model_potassium.predict(X_test)
accuracy_potassium = accuracy_score(y_test_potassium, y_pred_potassium)

print("Accuracy of Nitrogen Classification Model :", accuracy_nitrogen)
print("Accuracy of Phosphorus Classification Model :", accuracy_phosphorus)
print("Accuracy of Potassium Classification Model :", accuracy_potassium)
```

```python
X_test
```

```python
y_pred_nitrogen = model_nitrogen.predict(X_test)
```

```python
y_pred_nitrogen
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Generate confusion matrices
conf_matrix_nitrogen = confusion_matrix(y_test_nitrogen, y_pred_nitrogen)
conf_matrix_phosphorus = confusion_matrix(y_test_phosphorus, y_pred_phosphorus)
conf_matrix_potassium = confusion_matrix(y_test_potassium, y_pred_potassium)

# Plot confusion matrices
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.heatmap(conf_matrix_nitrogen, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Nitrogen")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.subplot(1, 3, 2)
sns.heatmap(conf_matrix_phosphorus, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Phosphorus")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.subplot(1, 3, 3)
sns.heatmap(conf_matrix_potassium, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Potassium")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.tight_layout()
plt.show()
```
