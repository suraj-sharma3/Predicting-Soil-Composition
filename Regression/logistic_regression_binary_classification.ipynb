# Linear Regression Curve / Equation ---> Given to Sigmoid Function ---> We get Logistic Regression Curve / Equation



```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(".\insurance_data.csv")
```

```python
df.head()
```

```python
plt.scatter(df.age, df.bought_insurance, marker='+', color="red")
```

### From the above scatter plot, we can see that people who are younger are less likely to buy an insurance while the older people are more ikely to buy an insurance (x axis values - bought insurance (1) or not (0) & y axis values - age of people)

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.9) # The training data has to be 2D that's why we are putting df[['age']] and not df.age
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test
```

```python
X_test.shape
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test)
```

```python
model.predict([[25]]) # Age - 25 years, a person who is 25 years old is not likely to buy an insurance
```

```python
model.predict([[56]]) # Age - 56 years, a person who is 50 years old is more likely to buy an insurance
```
